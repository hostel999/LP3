```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return (x+3)**2
def grad_f(x):
    return 2*(x+3)
```

```python
def gradient_descent(start_x,learning_ratio=0.1,max_iter=50,tolerance=1e-6):
    x =start_x
    history = [x]
    for _ in range(max_iter):
        grad = grad_f(x)
        x_new = x-learning_ratio* grad
        history.append(x_new)
        if abs(x_new-x)<tolerance:
            break
        x = x_new  
    return x,f(x),history
    
```

```python
min_x,min_y,steps = gradient_descent(start_x=2)
print("local Minima at x= ",min_x)
print("Minimal value at y: ",min_y)
```

```python
x_vals=np.linspace(-10,5,100)
y_vals = f(x_vals)
```

```python
plt.plot(x_vals,y_vals,label='y=(x+3)^2')
plt.scatter(steps,[f(x) for x in steps],color='red', label='Gradent descent steps')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gradient Descent Visualization')
plt.legend()
plt.show()
```
